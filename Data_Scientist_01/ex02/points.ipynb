```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data_train = pd.read_csv("/home/machouba/CC_42/Train_knight.csv")
data_test = pd.read_csv("/home/machouba/CC_42/Test_knight.csv")

group_train = data_train.groupby("knight").apply(lambda x: x)

fig = plt.figure(figsize=(25, 25))

plot = plt.subplot2grid((6, 5), (1, 0), fig=fig)

plot.scatter(data_test['Deflection'], data_test['Push'], alpha=0.5, label='Knight', color='#7fbf7f')
plot.set_xlabel('Deflection')
plot.set_ylabel('Push')
plot.legend()

plot = plt.subplot2grid((6, 5), (1, 1), fig=fig)

plot.scatter(data_test['Empowered'], data_test['Prescience'], alpha=0.5, label='Knight', color='#7fbf7f')
plot.set_xlabel('Empowered')
plot.set_ylabel('Prescience')
plot.legend()

plot = plt.subplot2grid((6, 5), (0, 1), fig=fig)

plt.scatter(group_train['Empowered']['Sith'], group_train['Prescience']['Sith'], label='Sith', color='red', alpha=0.3)
plt.scatter(group_train['Empowered']['Jedi'], group_train['Prescience']['Jedi'], label='Jedi', color='blue', alpha=0.3)
plot.set_xlabel('Empowered')   # x = Empowered
plot.set_ylabel('Prescience')  # y = Prescience
plot.legend()

plot = plt.subplot2grid((6, 5), (0, 0), fig=fig)

plot.scatter(group_train['Push']['Sith'], group_train['Deflection']['Sith'], label='Sith', color='red', alpha=0.3)
plot.scatter(group_train['Push']['Jedi'], group_train['Deflection']['Jedi'], label='Jedi', color='blue', alpha=0.3)
plot.set_xlabel('Push')
plot.set_ylabel('Deflection')
plot.legend()

plt.legend(loc="upper right")
plt.tight_layout()
plt.show()
```
