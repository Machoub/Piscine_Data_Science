```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("/home/machouba/CC_42/Test_knight.csv")
col = len(data.columns)
row = col // 5 + (1 if col % 5 else 0)

fig, axes = plt.subplots(nrows=row, ncols=5, figsize=(15, 15))
axes = axes.flatten()
for i, column in enumerate(data.columns):
    data[column].plot(kind="hist", ax=axes[i], title=column, color="lightgreen", bins=40)
    axes[i].set_title(f'{column}')
    axes[i].set_ylabel('')
    axes[i].legend(["Knight"])
plt.tight_layout()
plt.show()
```

```python
data_train = pd.read_csv("/home/machouba/CC_42/Train_knight.csv")
skills = [col for col in data_train.columns if col != 'knight']

fig, axes = plt.subplots(nrows=row, ncols=5, figsize=(15, 15))
axes = axes.flatten()
for i, skill in enumerate(skills):
    data_train.groupby("knight")[skill].plot(kind="hist", alpha=0.5, legend=True, bins=40, ax=axes[i])
    axes[i].set_title(f"{skill}")
    axes[i].set_ylabel('')
plt.tight_layout()
plt.show()
```
